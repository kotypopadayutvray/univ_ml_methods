```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
pd.set_option('display.max_columns', None)
```

```python
mushrooms = pd.read_csv('mushrooms.csv')
mushrooms.head()
```

```python
for column in mushrooms.columns:
    le = LabelEncoder()
    mushrooms[column] = le.fit_transform(mushrooms[column])
mushrooms.head()
```

Подготовка данных:

```python
X = mushrooms.drop(columns = ['class'])
y = mushrooms['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

Построение модели

Ссылки на документацию:
* https://www.tensorflow.org/learn
* https://www.tensorflow.org/guide/keras/sequential_model

```python
model = keras.Sequential(
    [
        layers.Dense(16, activation="relu"),
        layers.Dense(1, activation="sigmoid"),
    ]
)
model.compile(loss='binary_crossentropy')
```

```python
model.fit(X_train.to_numpy(), y_train.to_numpy(), batch_size=32, epochs=450)
```

```python
predictions = model.predict_classes(X_test.to_numpy()).flatten()
```

Выполняем оценку качества полученной модели с помощью [точности](https://scikit-learn.org/stable/modules/model_evaluation.html#accuracy-score) и ROC AUC (https://developers.google.com/machine-learning/crash-course/classification/roc-and-auc, https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_auc_score.html#sklearn.metrics.roc_auc_score)

```python
accuracy_score(y_test.to_numpy(), predictions.flatten())
```

```python
roc_auc_score(y_test.to_numpy(), predictions.flatten())
```
